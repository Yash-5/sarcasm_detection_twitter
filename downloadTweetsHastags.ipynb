```python
hashtags = ["#sad", "#depressed", "#hate", "#unlucky", "#terrible", "#lonely", "#sadness", "#worthless", "#empty", "#anxiety"]
```

```python
limit = 200000
write_limit = 5000
out_filename = "NegativeTweets.txt"
outfile = open(out_filename, "a")
all_tweets = set()
written_tweets = 0
prev_min_id = {}
i = 0
delimiter = "|||||"

while written_tweets + len(all_tweets) < limit:
    curr_tag = hashtags[i % len(hashtags)]
    try:
        if curr_tag in prev_min_id:
            search = twitter.search(q=curr_tag, lang='en', max_id=prev_min_id[curr_tag], count=100)
        else:
            search = twitter.search(q=curr_tag, lang='en', count=100)
    except:
        time.sleep(60)
        continue

    tweets = search["statuses"]

    try:
        min_id = str(min([int(x["id"]) for x in tweets]))
        prev_min_id[curr_tag] = min_id
    except:
        pass
    
    for x in tweets:
        all_tweets.add(x["text"])
    
    if len(all_tweets) > write_limit:
        for x in all_tweets:
            outfile.write(x + delimiter)
        written_tweets += len(all_tweets)
        all_tweets = set()
    i += 1
    print(written_tweets + len(all_tweets), end=" ")
```
